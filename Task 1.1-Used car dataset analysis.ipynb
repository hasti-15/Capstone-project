```python
path2=r'C:\Users\hasshah\Documents\Capstone3_UsedCarsData (1)\Capstone3_UsedCarsData\Sales_Status.csv'
sales=pd.read_csv(path2)
sales
```

```python
merged_data2=merged_data.merge(sales,on='Sales_ID',how='outer')
merged_data2.head()
def chec(x):
    if x=='Y':
        x=1
        return x
    else:
        x=0
        return x
        
df=merged_data2['Sold'].apply(chec)
merged_data2['Sold2']=df
merged_data2.head()

brand_data=merged_data2.groupby('Brand').agg({'Sold2':'sum'})
brand_data.plot(kind='bar')
```

Importing matplotlib and seaborn for visualization 
Constructing heat map to understand which factors are important in deciding car's selling price

```python
import matplotlib.pyplot as plt
import seaborn as sn
sn.heatmap(corr_matrix, annot=True, cmap="YlGnBu")
plt.show()
```
