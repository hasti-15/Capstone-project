```python
plt.scatter(car['km_Driven'],car['Selling_Price'])
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.title('Price Vs Kilometers Driven')
```

```python
mx = sns.factorplot(y="Selling_Price", x="Seats", data=car,kind="box",size=12 ,aspect=1)
mx.set(ylabel='Selling_Price')
mx.set(xlabel= 'Seats')
plt.title('Price Vs Seats');
```

```python
mx = sns.factorplot(x="Selling_Price", y="Brand", data=car,kind="box",size=12 ,aspect=1)
mx.set(xlabel='Old Price/New Price')
mx.set(ylabel= 'Brand')
plt.title('Brand Vs  Price');
```

```python
mx1 = sns.factorplot(x="Selling_Price", y="Year", data=car,kind="box",size=12 ,aspect=1)
mx1.set(xlabel='Selling Price')
mx1.set(ylabel= 'Year')
plt.title('Year Vs Price');
```
